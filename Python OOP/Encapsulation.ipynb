##### Encapsulation
###### Protecting data inside Class
###### Private properties can NOT be accessed directly outside the class. To access it, use GETTER or SETTER methods. 

```python
# Create Private class property named 
class Person:
    def __init__(self, name, age):
        self.name= name
        self.__age= age  # Private property
p1= Person("Emil",25)
print(p1.name)
# print(p1.__age) # This will cause ERROR
```

```python
# GETTER method
class Person:
    def __init__(self, name, age):
        self.name= name
        self.__age= age
    def getAge(self):   # GETTER
        return self.__age
p1= Person("Deepak",26)
print(p1.name)
print(p1.getAge())
```

```python
# SETTER method
class Person:
    def __init__(self, name):
        self.name= name
    def getName(self):
        return self.name
    def setName(self, t):
        self.name= t
p1= Person("Ramesh")
print("Before Setter:",p1.name)
p1.setName("Suresh")
print("After Setter:",p1.name)
```

```python
# EXAMPLE
class Student:
    def __init__(self, name):
        self.name= name
        self.grade= 0
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.grade= grade
        else:
            print("Grade must be between 0 and 100")
```
