```python
class Person:
    def __init__(self, name):
        self.name= name
    def printname(self):
        print("Hi there, I'm",self.name,"!")
p1= Person("Harinath")
p1.printname()
```

##### Methods with Parameters

```python
class Calculation:
    def addition(self,a,b):
        return a+b
    def multiplication(self,a,b):
        return a*b
c1= Calculation()
print(c1.addition(2,9))
print(c1.multiplication(2,9))
```

##### Methods accessing Properties

```python
class Person:
    def __init__(self, name, age):
        self.name= name
        self.age= age
    def getPerson(self):
        return f"{self.name} is {self.age} year old!"
p1= Person("Cristin",26)
print(p1.getPerson())
```

##### Methods modifying properties

```python
class Birthday:
    def __init__(self, age):
        self.age= age
    def printBirthday(self):
        self.age += 1
        print(f"Hi, you are {self.age} year old now! Happy Birthday🎉")
b1= Birthday(27)
b1.printBirthday()
```

##### The__str__() method
###### Controls what is returned when object is printed

```python
class OnePiece:
    def __init__(self, name):
        self.name= name
    def __str__(self):
        return f"{self.name} is in OnePiece World!"
o1= OnePiece("Luffy")
print(o1)
```

##### Delete Method

```python
class OnePiece:
    def __init__(self, name):
        self.name= name
    def strength(self):
        print(f"Roger's strength is {self.name}!")
o1= OnePiece("Haki")

del o1.strength # Throws an AttributeError
o1.strength()
```
