##### To access the inner class, create an object of the outer class, and then create an object of the inner class

```python
class Outer:
    def __init__(self):
        self.name = "Outer"
    class Inner:
        def __init__(self):
            self.name= "Inner"
        def display(self):
            print("Hello from Inner Class")
o1= Outer()
i1= o1.Inner()
i1.display()
```

##### Access Outer class from Inner class

```python
class Outer:
    def __init__(self):
        self.name= "Outer"
    def displayOuter(self):
        print("Hello from Outer Class")
    class Inner:
        def __init__(self, temp):
            self.temp= temp
        def displayInner(self):
            print(f"Hello from Inner Class: {self.temp.name}")
o1= Outer()
i1= o1.Inner(o1)
i1.displayInner()
```

```python
class Car:
    def __init__(self, brand, model):
        self.brand= brand
        self.model= model
        self.engine= self.Engine()  # self.Engine() creates object of Inner Class Engine
    class Engine:
        def __init__(self):
            self.status= "Off"  # it becomes car.engine.status = "Off"
        def start(self):
            self.status= "Running"
            print("Engine is Running...")
        def stop(self):
            self.stop= "Stop"
            print("Engine is Stopped.")
    def drive(self):
        if self.engine.status == "Running":
            print(f"Driving the {self.brand} {self.model}")
        else:
            print("Start the engine first!")
car= Car("Toyota", "Corolla")
car.drive()
car.engine.start()
car.drive()
```
