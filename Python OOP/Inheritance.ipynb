```python
# Parent Class
class Person:
    def __init__(self, fname, lname):
        self.fname= fname
        self.lname= lname
    def printname(self):
        print(self.fname,self.lname)
p1= Person("Rahul","Kumar")
p1.printname()

# Child Class
class Student(Person):
    pass
s1= Student("Priya","Sharma")
s1.printname()  # Now 'Student' class inherits all methods & attributes from Parent class 'Person'
```

##### Add the__init__() function
###### Child class__init__() function will override Parent class__init__() function. If we want to use parent class__init__() function, we need use ParentClass.__init__() function
###### with super(), never pass self as parameter, self is automatically passes.

```python
class Person:
    def __init__(self, name):
        self.name= name
        print(f"From Parent Class:{self.name}")

p1= Person("Pramod")
print("---------------------------------------------------")
class Employee(Person):
    def __init__(self, age):
        self.age= age
        print(f"From Child Class:{self.age}")
        Person.__init__(self, name="Pragati")
e1= Employee(26)
print("---------------------------------------------------")
class Student(Person):
    def __init__(self, document):
        self.document= document
        print(f"From Student Class:{self.document}")
        super().__init__("Prakash")
s1= Student("Aadhar")
```

##### Add Properties

```python
class Person:
    def __init__(self, fname, lname):
        self.fname= fname
        self.lname= lname
    def printname(self):
        print(self.fname,self.lname)

class Student(Person):
    def __init__(self, fname, lname, year):
        super.__init__(fname, lname)
        self.graduationyear= year

s1= Student("John","Doe",29)
print(s1.graduationyear)
```
